```python
from sklearn.cluster import k_means_ as k
from sklearn.datasets import make_blobs
```

```python
n_clusters = 3
```

```python
centers = [[2, 2], [6, 4], [3, 7]]
X, y = make_blobs(centers=centers, n_samples=3000, cluster_std=1)
```

```python
km = k.KMeans(n_clusters=n_clusters)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
from sklearn.utils.extmath import row_norms
from sklearn.utils import check_random_state
x_squared_norms = row_norms(X, squared=True)
random_state = check_random_state(14)
centroids = X[random_state.randint(0, X.shape[0], n_clusters)]

plt.figure(figsize=(20,20))
plt.scatter(X[:,0], X[:,1], c=y, s=40)
plt.scatter(centroids[:,0], centroids[:,1], c='red', s=100)
```

```python
import numpy as np

learning_rate = 1

all_centroids = []
centroids = X[random_state.randint(0, X.shape[0], n_clusters)]
all_centroids.append(centroids.copy())
counts = np.zeros((n_clusters,))
for iteration in range(5):
    previous_centroids = centroids.copy()
    distances = np.zeros((X.shape[0],), dtype=np.float64)
    inertia, squared_diff = k._mini_batch_step(X, x_squared_norms, centroids, counts,
centroids, False,
distances, random_reassign=False,
random_state=None, reassignment_ratio=.01,
verbose=False)
    centroids = (learning_rate * centroids) + ((1 - learning_rate) * previous_centroids)
    all_centroids.append(centroids.copy())
all_centroids[0], all_centroids[-1]
```

```python
labels, inertia = k._labels_inertia(X, x_squared_norms, centroids)
labels
```

```python
import numpy as np
all_centroids = np.array(all_centroids)
plt.figure(figsize=(40,40))
colors = ['red', 'green', 'blue']
markers = ['o', "o", "o"]
for centroid_index in range(len(all_centroids[0])):
    w = labels == centroid_index
    plt.scatter(X[w,0], X[w,1], c=colors[centroid_index], s=40, alpha=0.2)
for centroid_index in range(len(all_centroids[0])):
    #plt.plot(all_centroids[[0,-1],centroid_index,0], all_centroids[[0,-1],centroid_index,1], c='k', linewidth=10)
    #plt.plot(all_centroids[[0,-1],centroid_index,0], all_centroids[[0,-1],centroid_index,1], c='w', linewidth=5)
    dx, dy = (all_centroids[-1,centroid_index] - all_centroids[0,centroid_index])
    dx *= 0.7
    dy *=0.7
    plt.arrow(all_centroids[0,centroid_index,0], all_centroids[0,centroid_index,1], dx, dy,
             head_width=0.1, head_length=0.1, fc='k', ec='k', linewidth=5)
    plt.arrow(all_centroids[0,centroid_index,0], all_centroids[0,centroid_index,1], dx, dy,
             head_width=0.1, head_length=0.1, fc='w', ec='w', linewidth=3)
    plt.scatter(all_centroids[0,centroid_index,0], all_centroids[0,centroid_index,1], c=colors[centroid_index], marker="*", s=800)
    plt.scatter(all_centroids[-1,centroid_index,0], all_centroids[-1,centroid_index,1], c=colors[centroid_index], marker="^", s=800)
```
