```python
import numpy as np
from sklearn.datasets import make_blobs
```

```python
X_sphere, y = make_blobs(n_samples=1000, centers=[0,0])
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.scatter(X_sphere[:,0], X_sphere[:,1])
plt.axis([-10, 10, -10, 10])
```

```python
X_ellipse, y = make_blobs(n_samples=1000, centers=[0, 0])
X_ellipse[:,0] *= 2.5
X_ellipse[:,1] *= 1.5
plt.scatter(X_ellipse[:,0], X_ellipse[:,1])
plt.axis([-10, 10, -10, 10])
```

```python
centers = np.array([[0, 0], [-2.5, 2.5], [-5, 5], [-2.5, -2.5], [-1, -5], [0, -8], [3, 0], [4, 2]])
X_nonsmooth, y = make_blobs(n_samples=1000, centers=centers)
#X_nonsmooth[:,0] *= 2.5
#X_nonsmooth[:,1] *= 1.5
plt.scatter(X_nonsmooth[:,0], X_nonsmooth[:,1])
plt.axis([-10, 10, -10, 10])
```

```python
X_ellipse2 = X_ellipse.copy()
X_nonsmooth2 = X_nonsmooth.copy()
X_ellipse2[:,0] += 15
X_nonsmooth2[:,0] += 30
```

```python
plt.figure(figsize=(60, 20))

plt.scatter(X_sphere[:,0], X_sphere[:,1], color='b')
plt.scatter(X_ellipse2[:,0], X_ellipse2[:,1], color='r')
plt.scatter(X_nonsmooth2[:,0], X_nonsmooth2[:,1], color='g')
plt.axis([-5, 40, -10, 10])
```
