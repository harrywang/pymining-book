```python
import numpy as np
x = np.arange(-6, 6, 0.001)
```

```python
import math
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
y = np.array([sigmoid(xx) for xx in x])
```

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(50, 35))
plt.plot(x, y)
plt.plot(x, np.zeros(x.shape) + 0.5, color='r')
```
